```python
#Opens file
file_name = 'federalist_papers.txt'
try:
    federalist_papers = open(file_name, 'r')
    text = federalist_papers.read().lower()
except IOError:
    print("The file could not be opened! Make sure it exists!")
except:
    print("An file reading error has occured!")

punc = ['!','.', ',', ':', ';', '?', '-', '\n']

for puncmarc in punc:
    text = text.replace(puncmarc,' ')

word_list = text.split(" ")
```

```python
#opens stopwords.txt file
try: 
    stop_words = open('stopwords.txt','r')
    all_stop_words = stop_words.read().lower().strip()
except IOerror:
    print("The file could not be opened! Make sure it exists!")
except:
    print("A file reading error has occured.")

#turns txt file in list
stoplist = all_stop_words.split(',')
newwords = []
#removes stopwords and empty elements in the list. 
for word in word_list:
    if word == '':
        pass
    elif word not in stoplist:
        newwords.append(word)

        
```

```python
#Turns list into dicitonary with values that equal the freq. of that word
key_dict = {}

for word in newwords:
    if word in key_dict:
        key_dict[word] = key_dict[word] + 1
    else:
        key_dict[word] = 1

#Searches for user input in the dictionary and outputs freq. 
        
try:
    keyword = input("Enter Keyword:").strip().lower()
except: 
    print ("Input Failure")

if keyword in key_dict:
    print("The word " + keyword + " appears in the Federalist papers " + str(key_dict[keyword])  + " times.")
else:
    print(keyword + " is not in the Federalist Papers.")
```
